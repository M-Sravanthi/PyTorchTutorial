### Shape errors - one of the main errors in matrix multiplication
1. Inner dimesions must match                                                 
* '(3, 2) @ (3, 2)' will not work                                             
* '(2, 3) @ (3, 2)' will work                                                 
* '(3, 2) @ (2, 3)' will work                
* '@' is a matrix multiplication operator - but better to use 'matmul'
* The resulting matrix has the shape of outer dimensions:
* (2, 3) @ (3, 2) --> (2, 2)
* (3, 2) @ (2, 3) --> (3, 3)

```python
t @ t
```

```python
#torch.matmul(torch.rand(3,2),torch.rand(3,2)) #gives an error - shape error
```

```python
torch.matmul(torch.rand(2,3),torch.rand(3,2)) # output --> (2,2)
```

```python
torch.matmul(torch.rand(3,2),torch.rand(2,3)) # output --> (3,3)
```

### Reshaping, stacking, squeezing and unsqueezing Tensor
* Reshapeing - reshapes an input tensor to a defined shape                   
* View - Return a view of an input tensor of certain shape but keep the same memory as the original tensor

* Squeeze - removes all '1' dimensions from a torch.tensor

## PyTorch tensors and NumPy

NumPy is a popular scientific python numerical computing library
And bcoz of this PyTorch has functionality to interact with it

* Data in NumPy, want in PyTorch tensor - 'torch.from_numpy(ndarray)'

* PyTorch tensor -> NumPy -> torch.Tensor.numpy()

```python
# NumPy array to tensor
import torch
import numpy as np

array = np.arange(1.0,8.0)
tensor = torch.from_numpy(array) #Note: when converting from numpy to torch tensor, pytorch reflects numpy's default data type unless specified otherwise
array,tensor

#the default datatype of pytorch is float32 while that of numpy is float64
```

```python
array.dtype
```

```python
# Change the value of array, what will this do to 'tensor'?
array = array+1
array,tensor #they dont have shared memory
```

```python
# Tensor to NumPy array
tensor = torch.ones(7)
numpy_tensor = tensor.numpy()
tensor, numpy_tensor # the numpy array will now have default datatype of torch tensor - float32
```

```python
#Change the tensor - what happens to numpy tensor ?

tensor = tensor*2
tensor, numpy_tensor
```

## Reproducibility (trying to take random out of random)

In short how a neural network learns:

`Start with random numbers -> tensor operations -> update random numbers to
try and make them better representations of the data -> again -> again -> again`

To reduce the radomness in neural networks / pyTorch - we introduce the concept of `random seed`

Essentially what the random seed does is "flavour" the randomness.

```python
import torch

# Create two random tensors

random_tensor_A = torch.rand(3,4)
random_tensor_B = torch.rand(3,4)
print(random_tensor_A)
print(random_tensor_B)
print(random_tensor_A == random_tensor_B)
```

```python
# Let's make some random but reproducible tensors
import torch

# Set the random seed
RANDOM_SEED = 42

torch.manual_seed(RANDOM_SEED)
random_sensor_C = torch.rand(3,4)

torch.manual_seed(RANDOM_SEED)
random_sensor_D = torch.rand(3,4)

print(random_sensor_C)
print(random_sensor_D)
print(random_sensor_D == random_sensor_C)
```
